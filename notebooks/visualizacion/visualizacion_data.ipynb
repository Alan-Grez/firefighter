# Visualización de datos

Este notebook sirve como punto de partida para explorar y visualizar los datos del repositorio.
Actualiza la ruta si trabajas con otro archivo.

```python
from pathlib import Path
import pandas as pd

data_path = Path("../../data_2020/data_2020.csv")
df = pd.read_csv(data_path)
df.head()
```

```python
df.dtypes
```

```python
import plotly.express as px

numeric_cols = df.select_dtypes("number").columns
if len(numeric_cols) > 0:
    col = numeric_cols[0]
    fig = px.histogram(df, x=col, nbins=50, title=f"Distribución de {col}")
    fig.show()
else:
    print("No hay columnas numéricas para visualizar.")
```
